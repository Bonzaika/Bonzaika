Для нашего исследования пригодится только столбец cityCenters_nearest, поэтому заполним пропуски значений медианой в зависимости от населенного пункта. С остальными данными можно работать по аналогии. Но для исследования они нам не нужны. 

```python
#Для начала изменим наименование столбца:
data.rename(columns = {'cityCenters_nearest':'city_centers_nearest'}, inplace=True)
print('Всего пропущенных значений в столбце =', data['city_centers_nearest'].isna().sum())

# loc_median - здесь сохраним медианы расстояний по населенным пунктам
loc_median = data.groupby('locality_name')['city_centers_nearest'].median().sort_values()
loc_median = loc_median.dropna()
#display(loc_median.head(30))
#print('Всего медиан =', len(loc_median))

# Функция для присваивания пропуску значения медианы
def insert (loc_dist):
    if loc_dist not in loc_median:
        return np.nan
    return loc_median[loc_dist]
data.loc[data['city_centers_nearest'].isna()==True, 'city_centers_nearest'] = data['locality_name'].apply(insert)
print('Всего пропущенных значений в столбце =', data['city_centers_nearest'].isna().sum())
```

### Посчитатаем и добавим в таблицу цену за квадратный метр, добавим столбцы

```python
# Цена квадратного метра:
data['price_square_meter'] = data['last_price'] /data['total_area']
# Создаем столбцы: День недели, месяц и год публикации объявления
data['day_exp'] = data['first_day_exposition'].dt.weekday
data['month_exp'] = data['first_day_exposition'].dt.month
data['year_exp'] = data['first_day_exposition'].dt.year
display(data[['day_exp', 'month_exp', 'year_exp']].sample(8))
# Cоотношение жилой и общей площади:
data['ratio_usable_area'] = data['living_area'] /data['total_area']
# Cоотношение площади кухни к общей:
data['ratio_kitchen_area'] = data['kitchen_area'] /data['total_area']
data.head()
```

```python
# Этаж квартиры; варианты — первый, последний, другой:
def floor_category(row):
    floors_total = row['floors_total']
    floor = row['floor']
    if floor ==1:
        return 'первый'
    if floor == floors_total:
        return 'последний'
    return 'другой' 
#категоризуем этажи с помощью функции:
data['floor_category'] = data.apply(floor_category, axis = 1)
display(data[['floor', 'floors_total', 'floor_category']].sort_values(by='floor_category',ascending=True).sample(20))
```

### Уберем редкие и выбивающиеся значения.

Нам необходимо проанализировать и отбросить значения - выбросы для слеюущих столбцов: days_exposition, ceiling_height, 
kitchen_area, 'total_area', 'last_price', 'rooms'  
В данных столбцах есть как сильно большие знчения, так и очень маленькие. 

1. days_exposition: max = 1,580.00 , min = 1.00. В этом столбце откинем выбивающиеся значения, так как мы не  
2. ceiling_height: min= 1.00 , max = 100.00  
3. total_area: min= 12.00 ,  max = 900.00  
4. last_price: min =12,190.00 , max = 763,000,000.00  
5. rooms: min = 0.00 , max = 9.00


```python
# Границы выбросов найдем с рассчитаем с помощью iqr
# Функция для подсчёта границ с учетом 1.5 размаха
def quartile_range(df, column):
    q1 = df[column].quantile(0.25)
    q3 = df[column].quantile(0.75)
    iqr = q3 - q1
    dfq = df.loc[(df[column] < q3 + 1.5*iqr) & (df[column] > q1 - 1.5*iqr), column]
    return dfq

# В новом датафрейме удалим выбросы:
df_new = data.copy()
list_columns = ['days_exposition', 'ceiling_height','kitchen_area']
for i in list_columns:
    df_new[i] = quartile_range(data, i)
    share_emissions = 1-(df_new[i].count()/data[i].count())
    print('Доля удаленных значений в',i, '{:.0%}:'.format(share_emissions))
```

**Вывод** 
Мы изучили разброс данных и удалили выбивающиеся значения. Сохранили в новом датафрейме.

Условно этажи разделили на: первый, средние и последний. Квартиры, расположенные на первом этаже обычно дешевле. Это может быть связано с более высоким уровнем шума и повышенной вероятностью проникновения в квартиру воров. Это правило работает для всех районов Питера.  
Последний этаж как правило, дешевле остальных ввиду возможных проблем с крышей и потолками, а также из-за вероятности поломки лифта. При этом в центре города последний этаж практически не влияет на стоимость квартиры, в отличие от остального города. Возможно, верхние этажи выглядят более предпочтительными в связи с видовыми характеристиками и наличием в центре объектов класса «лакшери». 

**Удаленность от центра** 

Это важный фактор, влияющий на стоимость жилья: квартира в центре города стоит дороже той, что расположена ближе к окраине. Здесь играет роль  престижность и благополучие района, его инфраструктура – транспортная развязка (близость станций метро и остановок транспорта), близость образовательных и медицинских учреждений, торговых, развлекательных и культурных центров и т.д. Немаловажное значение имеет также экологическая обстановка. В центре Питера самые дорогие квартиры в расположены в диаметре до 1 км. Стоимость таких квартир достигает 170000 руб. за кв. метр.
